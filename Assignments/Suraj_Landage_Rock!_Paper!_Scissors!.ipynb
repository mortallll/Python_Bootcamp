##Rock! Paper! Scissors!

Make a rock paper scissors game where users can play with the computer. Use random.randint() to get computer's response. Print how many times the user won, tied, or lost. 


We will demo this game on 20-Jul. Be creative!

```python
# Rock! Paper! Scissors!
import random

# Game instruction
print("Rock! Paper! Scissors! Game")
print("")
print("Winning Rules of the Rock paper scissor game as follows: \n"
								+"Rock  vs paper   -> paper wins \n"
								+"Rock  vs scissor -> Rock wins \n"
								+"paper vs scissor -> scissor wins \n")

while True:
	print("Enter choice \n Rock    >> 1 \n Paper   >> 2 \n scissor >> 3 \n")
	
	choice = int(input("User turn: ")) # take the input from user
	
	
	while choice > 3 or choice < 1:    # looping until user enter valid input
		choice = int(input("enter valid input: "))
		

	# initialize value of choice_name variable
	if choice == 1:
		choice_name = 'Rock'
	elif choice == 2:
		choice_name = 'paper'
	else:
		choice_name = 'scissor'
		
	# print user choice
	print("user choice is: " + choice_name)
	print("\nNow its computer turn.......")

  # Computer chooses any number from 1 to 3 
  # Using randint method
	comp_choice = random.randint(1, 3) 
	
	# looping until comp_choice value
	while comp_choice == choice:
		comp_choice = random.randint(1, 3)

	# initialize value of comp_choice_name
	if comp_choice == 1:
		comp_choice_name = 'Rock'
	elif comp_choice == 2:
		comp_choice_name = 'paper'
	else:
		comp_choice_name = 'scissor'
		
	print("Computer choice is: " + comp_choice_name + "\n")

	print(choice_name + " V/s " + comp_choice_name)

	# condition for winning
	if((choice == 1 and comp_choice == 2) or
	(choice == 2 and comp_choice ==1 )):
		print("paper wins => ", end = "")
		result = "paper"
		
	elif((choice == 1 and comp_choice == 3) or
		(choice == 3 and comp_choice == 1)):
		print("Rock wins =>", end = "")
		result = "Rock"
	else:
		print("scissor wins =>", end = "")
		result = "scissor"

	# Printing either user or computer wins
	if result == choice_name:
		print("<== User wins ==>")
	else:
		print("<== Computer wins ==>")
  		
	ans = input("\nDo you want to play again? (Y/N): ")


	# if user input n or N then condition is True
	if ans == 'n' or ans == 'N':
		break
	
print("\nThanks for playing")
```
