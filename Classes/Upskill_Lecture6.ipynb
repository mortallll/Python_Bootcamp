#### Raw Strings

Place an `r` before beginning of quotation mark of a string to make it a raw string. A *raw string* completely ignores all escape characters and prints the string **as-is**.

Raw strings are used to define regex strings since those tend to have a lot of backslashes that would otherwise need to have escaped using more backslashes.

```python
print('Captain America's Shield.')
```

```python
print('Captain America\'s Shield.')
```

```python
print("Captain America's Shield.")
```

```python
print(r'Captain America's Shield.') 
```

```python
print(r'Captain America\'s Shield.')  #raw string
```

```python
print(r'Hello \\')
```

Slicing similar to strings: `string_name[start:stop:step]`

**Note:** 1. Stop index is not included in slicing.

2. Negative value in step means string is reversed.

3. Slicing does not make changes to original string. Store the sliced string in a variable to save changes.

```python
spam = 'Hello World!'
```

```python
spam[0:5:2] #slicing; slices till one index before stop
```

```python
spam[:5] #starts from 0 index
```

```python
spam[5:] #ends at last character
```

```python
spam[::-1] #reverses entire string
```

```python
#inplace = False
ham = spam[:5]
print(spam)
print(ham)
```

### Membership operators with strings.
The `in` and `not in` operators can be used with strings just like with list values. It returns a boolean value i.e `True` or `False`.

```python
s = "Hello Ironman"
```

```python
print(s)
print("Stark" in s)
print("Ironman" in s)
print("Ironman" not in s)
print("Stark" not in s)
```

### String Methods



#### Case Manipulation
1.   `upper()`: Returns new string where all letters in original string have been converted to uppercase. Non-letter characters remain unchanged.
2.   `lower()`: Same as `upper()` except letters are converted to lowercase.
3.   `isupper()`: Returns boolean true if string has at least one letter and all letters are uppercase. Returns false otherwise.
4.   `islower()` Same as `isupper()` except checks if all letters are lowercase.

```python
t = 'Hello Guys!'
```

```python
#upper() and lower()
print(t)
a = t.upper()
b = t.lower()
print(t)
print(a)
print(b)
```

```python
print(t)
```

```python
t.upper()
```

```python
print(t)
```

```python
a = 'HELLO'
b = 'WORLD'
```

```python
#isupper and islower()
print(a,a.isupper())
print(b, b.islower())
print(t,t.isupper(),t.islower())

c = '123'
print(c,c.isupper(),c.islower())

d = 'I <3 PYTHON!'
print(d, d.isupper(),d.islower())
```

#### isX()
Like `isupper()` and `islower()`, there are many validation methods that return boolean values. These are:
5.   `isalpha()`: Returns True if string only consists of **letters** and is not blank.
6.   `isalnum()`: Returns True if string consists only of **letters** and **numbers** and is not blank.
7.   `isdecimal()`: Returns True if string consists of only **numeric** characters and is not blank.
8.   `isspace()`: Returns True if the string only consists of **spaces, tabs and new-lines** and is not blank.
9.   `istitle()`: Returns True if the string consists of words that begin with an uppercase letter followed by only lowercase letters.



```python
#isX()
s = "HELLO WORLD"
print(s,s.isalpha())
print('Lotus123','Lotus123'.isalnum())
print(c, c.isdecimal())
print(spam,spam.istitle())
print("","".isalpha(), "".isalnum(), "".isdecimal(), "".isspace(), "".istitle())
```

```python
print('\n\t \n \t\t', '\n\t \n \t\t'.isspace())
```

#### Check start and end
10.  `startswith()`: Returns True if argument string present at starting of string on which method is called on. Returns false otherwise.
11.  `endswith()`: Same as `startswith()` except looks for argument substring at the end of string.

```python
spam = 'Hello World!'
```

```python
print(spam.startswith('Hello'), spam.endswith('World!'))
print(spam.startswith('hello'),spam.endswith('!dlroW'))
```

#### Join and Split
12.  `join()`: Used to join a list of strings into a single string value. Gets called on a string used to join, gets passed a list of strings that need to be joined and a final single string is returned.
13.  `split()`: Opposite of `join()`. It is called on a string, returns a list of strings and argument is used as a separator.


```python
#join
LISTJOINED = "~".join(LIST)
print(LIST,LISTJOINED)

#split
STRINGSPLIT = STRING.split(" ")
print(STRING, STRINGSPLIT)
```

```python
mylist = ['Hello', 'Guys', 'How', 'are', 'you']
print(" ".join(mylist))
```

#### Padding and justification
Methods that return padded version of string they are called on, with spaces to justify the text. It can take either one or two arguments. The first argument should be an integer that determines the total length of string after padding. The second optional argument determines what characters should be used for padding. Default is ` `.
14.  `rjust()`: Adds padding to left and shifts string to right.
15.  `ljust()` : Adds padding to right and shifts string to left.
16.  `center()`: Adds padding on both sides and shifts string to center.

```python
p = "Python Bootcamp"
len(p)
```

```python
#note: include size of string while passing padding size
print(p.rjust(30))
print(p.ljust(30))
print(p.center(30))
print(p.rjust(30,'#'))
print(p.ljust(30, '<'))
print(p.center(30, '='))
```

#### Stripping Strings
Sometimes you may want to strip off whitespace characters like spaces, tabs and newlines from left and/or right side of string. These stripping methods return a new string where whitespaces have been trimmed off. Optionally, an argument can be provided to trim those particular characters from the corners of string.
17.  `strip()`: Removes whitespace from both right and left side of a string.
18.  `rstrip()`: Removes whitespace from right side of string.
19.  `lstrip()`: Removes whitespace from left side of string.

**Note**: The order of characters provided as argument don't matter, the interpreter removes all matching characters until it encounters a character that doesn't match.

```python
new = "   Hello World!     "
print(new)
```

```python
print(new.strip())
```

```python
w = '\n  Natasha is a spy \n'
x = '\nShe has red hair\n'

print(w.strip() + '.')
print(w.lstrip())
print(w.rstrip())
print(w.strip() + '. ' + x.strip() + '.')
print(x.strip().rstrip('arih'))
print(x.strip().rstrip('hair'))
print(x.strip().rstrip('airh'))
print(x.strip().rstrip('ihar')) #does not depend on order
print(x.strip().lstrip('arih'))
```

```python
print(x.strip().lstrip('Seh'))
```

#### Find and count
20. `find()`: Searches for substring given as argument. The return value is the start index of the first match of the substring.
21. `count()`: Counts the number of substrings passed as argument in the string.

```python
y = 'What are you waiting here?'
print(y.find('a'))
print(y.count('a'))
```

### Readymade strings from String Module.



*   string.digits: 0123456789
*   string.punctuation: !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~
*   string.ascii_lowercase: abcdefghijklmnopqrstuvwxyz
*   string.ascii_uppercase: ABCDEFGHIJKLMNOPQRSTUVWXYZ



```python
#example of using strings defined in string module
import string

y = 'What do you want?!!&?'
print(y.rstrip(string.punctuation))
```

```python
y = '?!!What do you want?!!&?'
print(y.rstrip(string.punctuation))
print(y.lstrip(string.digits))
```
