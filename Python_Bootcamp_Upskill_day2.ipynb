Else Condition

```python
#check wheter a number is odd or even

number = int(input("Enter number: "))
if number % 2 == 0:
  print("Even Number")
else:
  print("Odd Number")
```

if-elif condition

```python
# You have to find th greater of 2 numbers.

x = 20
y = 10

if x>y:
  print(f"{x} is greater than {y}")
elif x == y:
  print("Both are equal")
else:
  print(f"{y} is greater than {x}")
```

Nested if statements

```python
number = int(input())
if number % 2 == 0:
  if number % 3 == 0:
    print("Divisible by 6")
  else:
    print("Not divisible by 6")
else:
  print("Not divisible by 6")
```

Loop control Statements

1. Break

```python
for i in range(1,10):
  print(i)
  if i==5:
    break
print("We are out of loop")
```

```python
for i in range(1,10):
  print(i)
  if i==5:
    break
  print("Hello")
print("We are out of loop")
```

```python
for i in range(1,10):
  print(i)
  if i==5:
    break
    print("Hello")
print("We are out of loop")
```

```python
for i in range(1,3):
  for j in range(1,5):
    if(j==3):
      break
print(j)
```

Continue Statement

```python
i = 0
while (i<8):
  i+=1
  if(i==5):
    continue
  print(i)
```

Pattern Programs

```python
rows = 6
for num in range(rows):
    for i in range(num):
        print("*", end=" ") # print number
    # line after each row to display pattern correctly

    print(" ")
```

```python
rows = 6
for num in range(rows):
    for i in range(num):
        print(num, end=" ") # print number
    # line after each row to display pattern correctly

    print(" ")
```
